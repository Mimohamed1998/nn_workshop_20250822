```python
import torch

x1 = torch.tensor([0.5], requires_grad=False)
x2 = torch.tensor([1.2], requires_grad=False)

w1 = torch.tensor([0.3], requires_grad=True)
w2 = torch.tensor([0.2], requires_grad=True)
w3 = torch.tensor([3.0], requires_grad=True)

# Forward pass
c = w1 * x1 + w2 * x2
c.retain_grad()  # retain gradient for intermediate tensor

d = torch.tanh(c)
d.retain_grad()

y = w3 * d
y.retain_grad()

t = 0.9
loss = 0.5 * (t - y) ** 2
```

```python
print("Value of d:", c.item())
print("Value of e:", d.item())
print("Value of y:", y.item())
print("Value of loss:", loss.item())
```

```python
# Backward pass
loss.backward()

# Display results
print("Gradient of loss w.r.t. y:", y.grad)
print("Gradient of loss w.r.t. e:", d.grad)
print("Gradient of loss w.r.t. d:", c.grad)

print("Gradient of loss w.r.t. w1:", w1.grad)
print("Gradient of loss w.r.t. w2:", w2.grad)
print("Gradient of loss w.r.t. w3:", w3.grad)
```
